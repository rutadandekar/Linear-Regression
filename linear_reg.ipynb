```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [2,3,4,5,7,9,16,25]
m = 2
c = 3
alpha = 0.0005
#y = [m*i+c for i in x]
y= map(lambda i:m*i+c,x)
print y
```

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
M = tf.Variable(4.0,tf.float32)
C = tf.Variable(2.0,tf.float32)
Y_hat = M*X + C
error = tf.reduce_sum(tf.square(Y-Y_hat))
opt = tf.train.GradientDescentOptimizer(alpha)
train = opt.minimize(error)
```

```python
ses = tf.Session()
ses.run(tf.global_variables_initializer())
for i in range(1000):
    ses.run(train,{X:x,Y:y})
    print ses.run([error,M,C],{X:x,Y:y})
ses.close()
```
